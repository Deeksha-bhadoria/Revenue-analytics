```python
import pandas as pd
import numpy as np
from pandasql import sqldf
import os
import warnings

warnings.filterwarnings("ignore")

# SQL shortcut — lets you write SQL on any DataFrame called 'df'
pysql = lambda q: sqldf(q, globals())

# Create output folder for Power BI files
os.makedirs("powerbi_data", exist_ok=True)
```

```python
df = pd.read_csv("data/claim_data.csv")

print("RAW SHAPE:", df.shape)
print("\nCOLUMNS:", df.columns.tolist())
print("\nFIRST 3 ROWS:")
display(df.head(3))
print("\nDATA TYPES:")
display(df.dtypes)
```

```python
# Step 1: Strip spaces from column names, make lowercase, replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Step 2: Parse date column properly
df["date_of_service"] = pd.to_datetime(df["date_of_service"], format="mixed")

# Step 3: Add useful time columns Power BI can use for time intelligence
df["month"]        = df["date_of_service"].dt.to_period("M").astype(str)  # e.g. 2024-05
df["month_name"]   = df["date_of_service"].dt.strftime("%b %Y")           # e.g. May 2024
df["month_num"]    = df["date_of_service"].dt.month                       # for sorting
df["year"]         = df["date_of_service"].dt.year
df["quarter"]      = df["date_of_service"].dt.quarter.apply(lambda x: f"Q{x}")

# Step 4: Create a calculated column — revenue gap per claim
df["revenue_gap"]  = df["billed_amount"] - df["paid_amount"]

# Step 5: Create denial flag (1 = denied, 0 = not denied)
df["is_denied"]    = (df["claim_status"] == "Denied").astype(int)

# Step 6: Create follow-up flag as numeric
df["followup_flag"] = (df["follow-up_required"].str.lower() == "yes").astype(int)

# Step 7: Strip whitespace from all text columns
text_cols = df.select_dtypes(include="object").columns
df[text_cols] = df[text_cols].apply(lambda col: col.str.strip())

# Verify
print("Data cleaned successfully.")
print(f"   Shape       : {df.shape}")
print(f"   Date range  : {df['date_of_service'].min().date()} → {df['date_of_service'].max().date()}")
print(f"   Null values : {df.isnull().sum().sum()}")
display(df.head(3))
```

```python
print("Claim Status Summary")
q = """
    SELECT
        claim_status,
        COUNT(*)                                                        AS total_claims,
        ROUND(SUM(billed_amount), 2)                                    AS total_billed,
        ROUND(SUM(allowed_amount), 2)                                   AS total_allowed,
        ROUND(SUM(paid_amount), 2)                                      AS total_paid,
        ROUND(SUM(billed_amount) - SUM(paid_amount), 2)                 AS revenue_gap,
        ROUND(AVG(billed_amount), 2)                                    AS avg_billed,
        ROUND(AVG(paid_amount), 2)                                      AS avg_paid,
        ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM df), 2)         AS pct_of_total
    FROM df
    GROUP BY claim_status
    ORDER BY total_claims DESC
"""
claim_status_summary = pysql(q)
display(claim_status_summary)
claim_status_summary.to_csv("powerbi_data/01_claim_status_summary.csv", index=False)
print("Exported: 01_claim_status_summary.csv")
```

```python
print("Insurance Type Summary")
q = """
    SELECT
        insurance_type,
        COUNT(*)                                                AS total_claims,
        ROUND(SUM(billed_amount), 2)                            AS total_billed,
        ROUND(SUM(allowed_amount), 2)                           AS total_allowed,
        ROUND(SUM(paid_amount), 2)                              AS total_paid,
        ROUND(SUM(billed_amount) - SUM(paid_amount), 2)         AS revenue_gap,
        ROUND(AVG(billed_amount), 2)                            AS avg_billed,
        ROUND(AVG(paid_amount), 2)                              AS avg_paid,
        ROUND(AVG(paid_amount) * 100.0 / AVG(billed_amount), 2) AS collection_rate_pct
    FROM df
    GROUP BY insurance_type
    ORDER BY total_billed DESC
"""
insurance_summary = pysql(q)
display(insurance_summary)
insurance_summary.to_csv("powerbi_data/02_insurance_summary.csv", index=False)
print("Exported: 02_insurance_summary.csv")
```

```python
print("Denial Reasons")
q = """
    SELECT
        reason_code,
        COUNT(*)                                                                    AS denied_count,
        ROUND(COUNT(*) * 100.0 / (SELECT COUNT(*) FROM df WHERE claim_status = 'Denied'), 2)
                                                                                    AS pct_of_denials,
        ROUND(SUM(billed_amount), 2)                                                AS billed_at_risk,
        ROUND(AVG(billed_amount), 2)                                                AS avg_billed
    FROM df
    WHERE claim_status = 'Denied'
    GROUP BY reason_code
    ORDER BY denied_count DESC
"""
denial_reasons = pysql(q)
display(denial_reasons)
denial_reasons.to_csv("powerbi_data/03_denial_reasons.csv", index=False)
print("Exported: 03_denial_reasons.csv")
```

```python
print("Monthly Trend")
q = """
    SELECT
        strftime('%Y-%m', date_of_service)      AS month,
        COUNT(*)                                AS total_claims,
        SUM(is_denied)                          AS denied_claims,
        ROUND(SUM(billed_amount), 2)            AS total_billed,
        ROUND(SUM(allowed_amount), 2)           AS total_allowed,
        ROUND(SUM(paid_amount), 2)              AS total_paid,
        ROUND(SUM(billed_amount) - SUM(paid_amount), 2) AS revenue_gap,
        ROUND(SUM(is_denied) * 100.0 / COUNT(*), 2)     AS denial_rate_pct
    FROM df
    GROUP BY month
    ORDER BY month
"""
monthly_trend = pysql(q)
display(monthly_trend)
monthly_trend.to_csv("powerbi_data/04_monthly_trend.csv", index=False)
print("Exported: 04_monthly_trend.csv")
```

```python
print("AR Status Risk")
q = """
    SELECT
        ar_status,
        COUNT(*)                                        AS total_claims,
        ROUND(SUM(billed_amount), 2)                    AS total_billed,
        ROUND(SUM(paid_amount), 2)                      AS total_paid,
        ROUND(SUM(billed_amount) - SUM(paid_amount), 2) AS revenue_at_risk,
        ROUND(AVG(billed_amount), 2)                    AS avg_billed,
        SUM(followup_flag)                              AS needs_followup
    FROM df
    GROUP BY ar_status
    ORDER BY revenue_at_risk DESC
"""
ar_status_summary = pysql(q)
display(ar_status_summary)
ar_status_summary.to_csv("powerbi_data/05_ar_status_summary.csv", index=False)
print("Exported: 05_ar_status_summary.csv")
```

```python
print("Top 10 Providers")
q = """
    SELECT
        provider_id,
        COUNT(*)                                        AS total_claims,
        ROUND(SUM(billed_amount), 2)                    AS total_billed,
        ROUND(SUM(paid_amount), 2)                      AS total_paid,
        ROUND(SUM(billed_amount) - SUM(paid_amount), 2) AS revenue_gap,
        SUM(is_denied)                                  AS total_denials,
        ROUND(SUM(is_denied) * 100.0 / COUNT(*), 2)    AS denial_rate_pct
    FROM df
    GROUP BY provider_id
    ORDER BY total_billed DESC
    LIMIT 10
"""
top_providers = pysql(q)
display(top_providers)
top_providers.to_csv("powerbi_data/06_top_providers.csv", index=False)
print("Exported: 06_top_providers.csv")
```

```python
print("Outcome vs Follow-up Cross Table")
q = """
    SELECT
        outcome,
        "follow-up_required",
        insurance_type,
        COUNT(*)                            AS total_claims,
        ROUND(SUM(billed_amount), 2)        AS total_billed,
        ROUND(SUM(paid_amount), 2)          AS total_paid
    FROM df
    GROUP BY outcome, "follow-up_required", insurance_type
    ORDER BY total_claims DESC
"""
outcome_detail = pysql(q)
display(outcome_detail)
outcome_detail.to_csv("powerbi_data/07_outcome_detail.csv", index=False)
print("Exported: 07_outcome_detail.csv")
```

```python
# Export the full cleaned dataset — Power BI's main table
df.to_csv("powerbi_data/00_claims_master.csv", index=False)

# Print executive summary 
total   = len(df)
billed  = df["billed_amount"].sum()
paid    = df["paid_amount"].sum()
gap     = billed - paid
denial  = df["is_denied"].mean() * 100
followup= df["followup_flag"].mean() * 100

print("=" * 55)
print("     HEALTHCARE CLAIMS — ANALYSIS SUMMARY")
print("=" * 55)
print(f"  Total Claims          : {total:,}")
print(f"  Total Billed          : ${billed:,.2f}")
print(f"  Total Paid            : ${paid:,.2f}")
print(f"  Revenue Gap           : ${gap:,.2f}")
print(f"  Overall Denial Rate   : {denial:.1f}%")
print(f"  Follow-up Required    : {followup:.1f}% of claims")
print("=" * 55)
print("\n📁 Files exported to: powerbi_data/")
print("   00_claims_master.csv")
print("   01_claim_status_summary.csv")
print("   02_insurance_summary.csv")
print("   03_denial_reasons.csv")
print("   04_monthly_trend.csv")
print("   05_ar_status_summary.csv")
print("   06_top_providers.csv")
print("   07_outcome_detail.csv")
```
